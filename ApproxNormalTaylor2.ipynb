## 標準正規分布のテイラー展開による近似

標準正規分布関数をテイラー展開を用いて近似してみましょう。テイラー展開は、関数を微分可能な点における導関数を利用して、その点周辺の多項式近似を導き出す方法です。

### 標準正規分布の確率密度関数のテイラー展開

標準正規分布の確率密度関数は、以下の式で表されます。

```
φ(x) = (1 / √(2π)) * exp(-x^2 / 2)
```

この関数を原点 (x = 0) においてテイラー展開すると、以下のようになります。

```
φ(x) ≈ 1 - x^2 / 2 + x^4 / 24 - x^6 / 720 + ...
```

### テイラー展開を用いた近似関数の作成

テイラー展開を用いた近似関数は、以下のコードで実装できます。

````
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.special import factorial

# 標準正規分布の確率密度関数
def standard_normal_pdf(x):
    return norm.pdf(x)

# テイラー展開に基づく多項式近似
def taylor_series_approximation(x, degree):
    sum = 0
    for n in range(degree + 1):
        term = ((-1)**n * x**(2*n)) / (factorial(n) * 2**n)
        sum += term
    return (1 / np.sqrt(2 * np.pi)) * sum

# xの範囲
x = np.linspace(-2, 2, 400)

# 標準正規分布
y_normal = standard_normal_pdf(x)

# プロットの準備
plt.figure(figsize=(10, 6))

# 標準正規分布をプロット
plt.plot(x, y_normal, label='Standard Normal Distribution', color='black')

# 多項式近似のプロット
degrees = [2, 4, 6, 8, 10]  # 多項式の次数
colors = ['blue', 'green', 'red', 'purple', 'orange']

for degree, color in zip(degrees, colors):
    y_taylor = taylor_series_approximation(x, degree)
    plt.plot(x, y_taylor, label=f'Taylor Series (degree={degree})', color=color)

# プロットの設定
plt.xlabel('x')
plt.ylabel('Density')
plt.title('Standard Normal Distribution and its Taylor Series Approximations')
plt.legend()
plt.grid(True)
plt.show()
````
