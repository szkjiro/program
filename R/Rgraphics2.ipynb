# Rによるグラフ作成いろいろ

まずは散布図のバリエーション，つまりplot関数の扱いから始める．
ここではRの用意するサンプルデータMASSのもつCars93を利用する．
しばらくは，このデータ利用で進めるので，attach関数により入力指定の簡略化もしておく．

参考文献：山田剛史ほか『統計データの視覚化』（オーム社）


```r
data(Cars93,package="MASS")
head(Cars93)
attach(Cars93)
```

いろいろなデータ列をもつが，ここではMPG.city（Mile Per Gallon，ガソリン1ガロンあたりの走行距離）の都市部Cityにおけるもの，AirBags（エアバッグの装備状況），Horsepower（エンジンの馬力），Cylinders（エンジンのシリンダー数つまり気筒数）のみ扱っていく．

いくつかグラフを描かせてみよう．
変数を一つだけしか与えていないところは，該当するものをカウントした値である．
各変数が質的変数であるか量的変数であるかの区別がつくだろうか．

```r
plot(AirBags)
plot(MPG.city)
plot(Cylinders,AirBags)
plot(Horsepower,MPG.city)
plot(AirBags,MPG.city)
plot(MPG.city,AirBags)
```

3つ以上の変数をデータフレイムの形で与えれば，散布図行列を描いてくれる．

```r
plot(data.frame(MPG.city,AirBags,Horsepower))
```

散布図はx軸の変数を説明変数，y軸の変数を目的変数として，xによりyを説明させるという考え方をする．
Rでは記号「｀」を利用して「y~x」のような書き方をする．

```r
plot(MPG.city~Horsepower)
plot(MPG.city~AirBags)
plot(AirBags~MPG.city)
```

3つのうち2つはすでに描かせたグラフに同じである．
3つ目だけ層別の点プロットだったものが密度プロットになっている．

問：実際にどれと同じであるか．また先に扱ったplot関数の描画のうち，見た目の変わるものはあるか．

## 点のパターン変更など

散布図のプロットを○以外にしたり，線種や色も変更できる．
まずはパターン変更してみよう．

```r
plot(MPG.city)
plot(MPG.city,type="l")
plot(MPG.city,type="b")
plot(MPG.city,type="o")
```

次は線の太さや種類を変更してみる．

```r
plot(MPG.city,type="b",lwd=2,pch=17)
plot(MPG.city,type="b",lty=2)
```
